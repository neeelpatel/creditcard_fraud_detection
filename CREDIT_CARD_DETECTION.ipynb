```python
# dependent(target) and independent variable
x=final_dataset.drop('Class',axis=1)
```

```python
y=final_dataset.Class
```

```python
#split the data into train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=10)
```

```python
print(x_train.shape)
print(x_test.shape)
```

Model training
Logistic regression


```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#accuracy check on training set
x_train_prediction=model.predict(x_train)
accuracy_score(x_train_prediction,y_train)
```

```python
#accuracy check on test set
y_predict = model.predict(x_test)
accuracy_score(y_test,y_predict)
```

```python
matrix = confusion_matrix(y_test,y_predict)
matrix
```

```python
#create dependent and independentvariable
x=data1.drop("Class",axis=1)
x
```

```python
y=data1.Class
y
```

```python
#model
classifier = {
    "IsolationForest ":IsolationForest(n_estimators=100,verbose=0,max_samples=len(x)),
    "local Outlier Factor":LocalOutlierFactor(n_neighbors=20)
    
}
```

```python
classifier.items()
```

```python
for i, (clf_name,clf) in enumerate(classifier.items()):
    if clf_name == "IsolationForest":
        y_pred = clf.fit_predict(x)
    else:
        clf.fit(x)
        y_pred = clf.predict(x)

        
    
    
    
```

```python
for i, (clf_name,clf) in enumerate(classifier.items()):
    #Fit the data and tag outliers
    if clf_name == "Local Outlier Factor":
        y_pred = clf.fit_predict(x)
        scores_prediction = clf.negative_outlier_factor_
    else:    
        clf.fit(x)
        scores_prediction = clf.decision_function(x)
        y_pred = clf.predict(x)
    #Reshape the prediction values to 0 for Valid transactions , 1 for Fraud transactions
    y_pred[y_pred == 1] = 0
    y_pred[y_pred == -1] = 1
    n_errors = (y_pred != y).sum()
    # Run Classification Metrics
    print("{}: {}".format(clf_name,n_errors))
    print("Accuracy Score :")
    print(accuracy_score(y,y_pred))
    print("Classification Report :")
    print(classification_report(y,y_pred))


     
 
```
